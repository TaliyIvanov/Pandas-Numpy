```python
#РЕШЕНИЕ
import pandas as pd
import  numpy as np

d = {'ВИП': 1,  
     'Все за 300': 2, 
     'Всё за 500': 3,  
     'Всё за 800': 4, 
     'Промо': 7, 
     'Корпоративный': 0, 
     'Всё за 250 (архив)': 5}
    
def solution(df):
    
    df['sms_price'] = df['tarif'].map(d)
    df.loc[df['city'] == 'Moscow', ['balance']]=df['balance'] - df['sms'] * df['sms_price']
    df = df.drop(['sms_price'], axis = 1)
    
    return df
```
