```python
import numpy as np
import pandas as pd
```

# Мультииндексы в колонках

```python
data = np.array([[50000, 16, 50000, 16],
                 [45000, 15, 45000, 15,],
                 [40000, 14, 40000, 14],
                 [55000, 15, 55000, 15],
                 [47000, 13, 47000, 13],
                 [45000, 12, 45000, 12],
                 [52000, 17, 52000, 17],
                 [43000, 15, 43000, 15],
                 [41000, 13, 41000, 13]])

index = [['Тестирование', 'Тестирование', 'Тестирование', 'Разработка', 'Разработка', 'Разработка', 
          'Аналитика', 'Аналитика', 'Аналитика'], 
         ['Начальник', 'Эксперт', 'Специалист', 'Начальник', 'Эксперт', 'Специалист', 'Начальник', 
          'Эксперт', 'Специалист']]

columns = [['Москва', 'Москва', 'Казань', 'Казань'], ['Зарплата', 'Уровень', 'Зарплата', 'Уровень']]

df = pd.DataFrame(data, index=index, columns=columns)
```

```python
df
```

```python
#Получим имена уровней
df.index.names
```

```python
#Их нет, поэтому зададим их
df.index.names = ['Отдел','Должность']
```

```python
df
```

```python
#Повторим тоже самое для индексов колонок
#Заново инициализируем датафрэйм
df
```

```python
df.columns.names = ['Город','Показатель']
```

```python
df
```

```python
#Теперь совместим и то и другое
#Снова инициализируем датафрэйм
df.index.names = ['Отдел','Должность']
df.columns.names = ['Город','Показатель']
```

```python
df
```

# Изменения индексов

- .swaplevel() - меняет индексы местами

```python
df.swaplevel(0, 1)
```

```python
#swaplevel может принимать так же и лейблы
df.swaplevel('Отдел', 'Должность')
```

- сортировка

```python
df.sort_index(level = 0)
```

```python
df.sort_index(level = 1)
```

- .reset_index()

```python
df.reset_index()
```
